```python
f = open("./pdata.json", 'rt', encoding="utf8")
```

```python
dataset = []
```

```python
for line in f:
    line = eval(line)
    dataset.append(line)
```

```python
dataset[0]
```

```python
oneStar = []
twoStar = []
threeStar = []
fourStar = []
fiveStar = []
```

```python
for d in dataset:
    if d['overall'] == 1:
        oneStar.append(d)
    if d['overall'] == 2:
        twoStar.append(d)
    if d['overall'] == 3:
        threeStar.append(d)
    if d['overall'] == 4:
        fourStar.append(d)
    if d['overall'] == 5:
        fiveStar.append(d)
```

```python
import matplotlib.pyplot as plt
```

```python
names = ['one', 'two', 'three', 'four', 'five']
values = [len(oneStar), len(twoStar), len(threeStar), len(fourStar), len(fiveStar)]
plt.bar(names, values)
plt.show()
```

```python
import numpy as np
sum = 0
for d in dataset:
    sum += d['overall']
print("The average rating of this dataset is " + str(sum/len(dataset)))
```

```python
user = set()
item = set()
for d in dataset:
    user.add(d['reviewerID'])
    item.add(d['asin'])
print("User, Item: " + str(len(user)) +", "+ str(len(item)))
```

```python
import string
punct = string.punctuation
```

```python
word_array = []
for d in dataset:
    t = d['reviewText']
    t = t.lower() # lowercase string
    t = [c for c in t if not (c in punct)] # non-punct characters
    t = ''.join(t) # convert back to string
    words = t.strip().split() # tokenizes
    for w in words:
        word_array.append(w)
```

```python
word_string = ' '.join(word_array)
```

```python
from wordcloud import WordCloud
wordcloud = WordCloud(width=600, height=500, margin=0).generate(word_string)
```

```python
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.margins(x=0, y=0)
plt.show()
```
